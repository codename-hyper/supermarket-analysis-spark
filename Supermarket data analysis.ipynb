```
from pyspark import SparkContext, SparkConf
conf = SparkConf().setAppName("practice")
sc = SparkContext.getOrCreate(conf)
```

```
file = sc.textFile("/FileStore/tables/supermarket_sales___Sheet1.csv")
```

```
header = file.first()
```

### How much does an individual branch earn

```
branch_earn = file.filter( lambda x : x != header ).map( lambda x : ( x.split(",")[1], float(x.split(",")[9]) ) ).reduceByKey( lambda x,y : x+y )
branch_earn.collect()
```

### What is the total amount spent by each gender

```
gender_spend = file.filter( lambda x : x != header ).map( lambda x : ( x.split(",")[4], float(x.split(",")[9]) ) ).reduceByKey( lambda x,y : x+y )
gender_spend.collect()
```

### Which product line sells the most

```
prod_line = file.filter( lambda x : x != header ).map( lambda x : ( x.split(",")[5], 1 ) ).reduceByKey( lambda x,y : x+y )
prod_line.collect()
```

### Which payment mode does customer prefers

```
pay_mode = file.filter( lambda x : x != header ).map( lambda x : ( x.split(",")[12], 1 ) ).reduceByKey( lambda x,y : x+y )
pay_mode.collect()
```

### Which time do customer prefer to visit the supermarket

```
prefered_time = file.filter( lambda x : x != header ).map( lambda x : (int(x.split(",")[11][:2]), 1) ).reduceByKey( lambda x,y : x+y )
prefered_time.collect()
```

### How many loyal customers per city

```
loyal_per_city = file.filter( lambda x : x != header ).map( lambda x : ( x.split(",")[2], x.split(",")[3] ) ).filter( lambda x : x[1] == "Member" ).map( lambda x : (x[0], 1) ).reduceByKey( lambda x,y : x+y )
loyal_per_city.collect()
```

### How much does supermarket earn per city

```
city_earn = file.filter( lambda x : x != header ).map( lambda x : ( x.split(",")[2], float(x.split(",")[9]) ) ).reduceByKey( lambda x,y : x+y )
city_earn.collect()
```

### Average rating per branch

```
avg_rating = file.filter( lambda x : x != header ).map( lambda x : (x.split(",")[1], ( float(x.split(",")[-1]), 1 ) ) ).reduceByKey( lambda x,y : ( x[0]+y[0], x[1]+y[1] ) ).map( lambda x : ( x[0], x[1][0]/x[1][1] ) )
avg_rating.collect()
```
